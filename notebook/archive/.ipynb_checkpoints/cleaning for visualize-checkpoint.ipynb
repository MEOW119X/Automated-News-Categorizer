```python
import pandas as pd
```

```python
df01 = pd.read_csv(r'C:\Users\nihen\Desktop\testing plotly dash\raw\01.2020 no label.csv')
df02 = pd.read_csv(r'C:\Users\nihen\Desktop\testing plotly dash\raw\02.2020 no label.csv')
```

```python
la01 = pd.read_csv(r'C:\Users\nihen\Desktop\testing plotly dash\cleaned\label 2020.01.csv')
la02 = pd.read_csv(r'C:\Users\nihen\Desktop\testing plotly dash\cleaned\label 2020.02.csv')
```

```python
df01['month'] = '01'
df02['month'] = '02'
```

```python
la01 = la01.rename(columns={'url':'label'})
la02 = la02.rename(columns={'url':'label'})
```

```python
df01 = pd.concat([df01, la01], axis=1)
df02 = pd.concat([df02, la02], axis=1)
```

```python
df01 = df01[['date', 'month', 'label', 'headline', 'snippet', 'url']]
df02 = df02[['date', 'month', 'label', 'headline', 'snippet', 'url']]
```

```python
df03 = df01.merge(df02, how='outer')
```
