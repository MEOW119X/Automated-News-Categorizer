```python
import time
import requests
import dateutil
import pandas as pd
```

```python
date = ['2020', '1']
```

```python
def send_request(date):
    '''Sends a request to the NYT Archive API for given date.'''
    base_url = 'https://api.nytimes.com/svc/archive/v1/'
    url = base_url + '/' + date[0] + '/' + date[1] + '.json?api-key=' + 'hZpfTt35C1OK7ibsJn8Ti1IXOzHCb0HU'
    response = requests.get(url).json()
    time.sleep(6)
    return response

def parse_response(response):
    '''Parses and returns response as pandas data frame.'''
    data = {'date': [],
        'headline': [],
        'snippet': [],
        'url': []}

    articles = response['response']['docs'] 
    for article in articles: # For each article, make sure it falls within our date range
        #date
        date = dateutil.parser.parse(article['pub_date']).date()
        data['date'].append(date)

        #headline
        data['headline'].append(article['headline']['main'])

        #snippet
        data['snippet'].append(article['snippet'])

        #web url
        data['url'].append(article['web_url'])
    return pd.DataFrame(data) 
```

```python
datex = send_request(date)
x = parse_response(datex)
```

```python
x
```

```python
x['label'] = x['url']
```

```python
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/(interactive)\/\d+\/\d+\/\d+\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/\d+\/\d+\/\d+\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/(slideshow)\/\d+\/\d+\/\d+\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/(interactive)\/\d+\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/(video)\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/www.nytimes.com\/)','', regex=True)
x['label'] = x['label'].str.replace(r'(https?:\/\/brandedplaylist.nytimes.com\/)','', regex=True)
```

```python
x['label'] = x['label'].str.replace(r'(\/.+)','', regex=True)
x['label'] = x['label'].str.replace(r'\s+','', regex=True)
```
